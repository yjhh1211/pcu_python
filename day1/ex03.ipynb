```python
import numpy as np
```

```python
t=np.arange(0,2*np.pi,0.01)

plt.figure(figsize=(10,6))
plt.plot(t,np.sin(t))
plt.grid()
plt.title('sine graph')
plt.xlabel('sec')
plt.ylabel('sin')
plt.show()
```

```python
t = np.arange(0,2*np.pi,0.01)

plt.figure(figsize=(10,6))
plt.plot(t,np.sin(t), lw=3, label='sin')
plt.plot(t,np.cos(t), 'r', label='cos')
plt.grid()
plt.legend()
plt.title('sine')
plt.xlabel('sec')
plt.ylabel('sin')
plt.show()
```
